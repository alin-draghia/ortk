```python
zero_size = ort.Size()
zero_size.width = 0;
zero_size.height = 0;
window_size = ort.Size()
window_size.width = 64
window_size.height = 128
window_step = ort.Size()
window_step.width = 8
window_step.height = 8
pyramidBuilder = ort.create_FloatPyramidBuilder(1.1, zero_size, zero_size)
scanner = ort.create_DenseImageScanner(window_size, window_step, zero_size)
featureExtractor = ort.create_HogFeatureExtractor()
classifier = ort.create_MockPersonClassifier()
nms = ort.create_PassThroughNms()
nms2 = ort.create_GroupRectanglesNms()
print pyramidBuilder
print scanner
print featureExtractor
print classifier
print nms
print nms2
```

```python
detector = ort.build_DetectorMT(pyramidBuilder, scanner, featureExtractor, classifier, nms2)
```

```python
image = cv2.imread(r'../datasets/INRIAPerson/Test/pos/crop001520.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
dets = ort.BoxVector()
conf = ort.DoubleVector()
detector.Detect(image, dets, conf, 0.0)
```

```python
print dets, len(dets)
print conf, len(conf)
for d in dets:
    x = d.x
    y = d.y
    w = d.width
    h = d.height
    cv2.rectangle(image, (x,y), (x+w,y+h), (255,0,0))
```
